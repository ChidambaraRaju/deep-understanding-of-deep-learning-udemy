# 09: Dropout Regularization Experiment

### ðŸŽ¯ Objective
This notebook tests the effectiveness of **Dropout Regularization**. We will use a custom "doughnut" dataset that is prone to overfitting and see if adding dropout layers helps the model generalize better to unseen data.

### ðŸ“š Key Concepts
- **The Doughnut Problem:** A non-linear classification task with concentric circles.
- **Dropout Rate:** The probability $p$ that any given neuron will be turned off.
- **Generalization Gap:** The difference between Training Accuracy and Test Accuracy. A smaller gap usually means better generalization.
- **Hyperparameter Sweep:** Systematically testing a range of dropout rates (0.0 to 0.9).

```python
# import libraries
import torch
import torch.nn as nn
import torch.nn.functional as F

import numpy as np
import matplotlib.pyplot as plt
from IPython import display
display.set_matplotlib_formats('svg')

from sklearn.model_selection import train_test_split
from torch.utils.data import DataLoader,TensorDataset
```

## 1. Create the Data (The Doughnuts)

We generate two classes of data:
1.  **Inner Circle (Blue):** Radius ~10.
2.  **Outer Circle (Black):** Radius ~15.



This data is non-linear (you can't separate rings with a straight line) and noisy, making it a good candidate for overfitting.

```python
# create data

nPerClust = 200

th = np.linspace(0,4*np.pi,nPerClust)
r1 = 10
r2 = 15

# generate data
a = [ r1*np.cos(th) + np.random.randn(nPerClust)*3 ,
      r1*np.sin(th) + np.random.randn(nPerClust) ]
b = [ r2*np.cos(th) + np.random.randn(nPerClust) ,
      r2*np.sin(th) + np.random.randn(nPerClust)*3 ]

# true labels
labels_np = np.vstack((np.zeros((nPerClust,1)),np.ones((nPerClust,1))))

# concatanate into a matrix
data_np = np.hstack((a,b)).T

# convert to a pytorch tensor
data = torch.tensor(data_np).float()
labels = torch.tensor(labels_np).float()

# show the data
fig = plt.figure(figsize=(5,5))
plt.plot(data[np.where(labels==0)[0],0],data[np.where(labels==0)[0],1],'bs')
plt.plot(data[np.where(labels==1)[0],0],data[np.where(labels==1)[0],1],'ko')
plt.title("The qwerties' doughnuts!")
plt.xlabel('qwerty dimension 1')
plt.ylabel('qwerty dimension 2')
plt.show()
```

## 2. Prepare DataLoaders
Standard splitting and batching process.

```python
# use scikitlearn to split the data
train_data,test_data, train_labels,test_labels = train_test_split(data, labels, test_size=.2)


# then convert them into PyTorch Datasets (note: already converted to tensors)
train_data = TensorDataset(train_data,train_labels)
test_data  = TensorDataset(test_data,test_labels)


# finally, translate into dataloader objects
batchsize    = 16 #int(train_data.tensors[0].shape[0]/4) -- Hard-coding is better to avoid huge batches!
train_loader = DataLoader(train_data,batch_size=batchsize,shuffle=True)
test_loader  = DataLoader(test_data,batch_size=test_data.tensors[0].shape[0]) # how big should these batches be??
```

## 3. Create the Model Class

This class is designed to be flexible. It takes a `dropoutRate` as an input argument, allowing us to easily create models with different levels of regularization.

**Architecture:**
* **Input:** 2 Features
* **Hidden:** 128 Units -> ReLU -> **Dropout**
* **Hidden:** 128 Units -> ReLU -> **Dropout**
* **Output:** 1 Unit (Binary Classification)

**Key Code:** `F.dropout(x, p=self.dr, training=self.training)`. Passing `self.training` ensures dropout is automatically disabled when we call `model.eval()`.

```python
class theModelClass(nn.Module):
  def __init__(self,dropoutRate):
    super().__init__()

    ### layers
    self.input  = nn.Linear(  2,128)
    self.hidden = nn.Linear(128,128)
    self.output = nn.Linear(128, 1)

    # parameters
    self.dr = dropoutRate

  # forward pass
  def forward(self,x):

    # pass the data through the input layer
    x = F.relu( self.input(x) )

    # dropout after input layer
    x = F.dropout(x,p=self.dr,training=self.training) # training=self.training means to turn off during eval mode

    # pass the data through the hidden layer
    x = F.relu( self.hidden(x) )

    # dropout after hidden layer
    x = F.dropout(x,p=self.dr,training=self.training) # training=self.training means to turn off during eval mode

    # output layer
    x = self.output(x)
    # no dropout here!!
    return x
```

## 4. Functions for Creating and Training

We refactor the logic into functions to support our experiment loop.

**Loss Function:** `nn.BCEWithLogitsLoss()`. This combines a Sigmoid layer and the BCELoss into one single class. It is more numerically stable than using `Sigmoid` + `BCELoss` separately.

```python
def createANewModel(dropoutrate):

  # grab an instance of the model class
  ANNQC = theModelClass(dropoutrate)

  # loss function
  lossfun = nn.BCEWithLogitsLoss()

  # optimizer
  optimizer = torch.optim.SGD(ANNQC.parameters(),lr=.002)

  return ANNQC,lossfun,optimizer
```

```python
# train the model

# training parameter
numepochs = 1000

# NOTE: this time, the model, lossfun, and optimizer are inputs into the function!
def trainTheModel(ANNQC,lossfun,optimizer):

  # initialize accuracies as empties (not storing losses here)
  trainAcc = []
  testAcc  = []


  # loop over epochs
  for epochi in range(numepochs):

    # switch training mode on (is this the right place for this code??)
    # YES! We need to make sure dropout is active for the training loop.
    ANNQC.train()

    # loop over training data batches
    batchAcc = []
    for X,y in train_loader:

      # forward pass and loss
      yHat = ANNQC(X)
      loss = lossfun(yHat,y)

      # backprop
      optimizer.zero_grad()
      loss.backward()
      optimizer.step()

      # compute training accuracy just for this batch
      batchAcc.append( 100*torch.mean(((yHat>0)==y).float()).item() )
    # end of batch loop...

    # now that we've trained through the batches, get their average training accuracy
    trainAcc.append( np.mean(batchAcc) )

    # test accuracy
    ANNQC.eval() # switch training off (DISABLE DROPOUT for testing)
    X,y = next(iter(test_loader)) # extract X,y from test dataloader
    yHat = ANNQC(X)
    testAcc.append( 100*torch.mean(((yHat>0)==y).float()).item() )

  # function output
  return trainAcc,testAcc
```

## 5. Test the Model (Single Run)
Running it once with `dropoutrate = 0.0` (No dropout) to establish a baseline.

```python
# create a model
dropoutrate = .0
ANNQC,lossfun,optimizer = createANewModel(dropoutrate)

# train the model (note the inputs!)
trainAcc,testAcc = trainTheModel(ANNQC,lossfun,optimizer)
```

```python
# create a 1D smoothing filter
# This helps make the noisy accuracy plots easier to read
def smooth(x,k=5):
  return np.convolve(x,np.ones(k)/k,mode='same')
```

```python
# plot the results
fig = plt.figure(figsize=(10,5))

plt.plot(smooth(trainAcc),'bs-')
plt.plot(smooth(testAcc),'ro-')
plt.xlabel('Epochs')
plt.ylabel('Accuracy (%)')
plt.legend(['Train','Test'])
plt.title('Dropout rate = %g'%dropoutrate)

plt.show()
```

## 6. The Experiment: Varying Dropout Rates

We will now loop through dropout rates from 0.0 to 0.9.


```python
# run an experiment (note: takes ~3.5 mins)

dropoutRates = np.arange(10)/10
results = np.zeros((len(dropoutRates),2))

for di in range(len(dropoutRates)):

  # create and train the model
  ANNQC,lossfun,optimizer = createANewModel(dropoutRates[di])
  trainAcc,testAcc = trainTheModel(ANNQC,lossfun,optimizer)

  # store accuracies from last 100 epochs (to get a stable average)
  results[di,0] = np.mean(trainAcc[-100:])
  results[di,1] = np.mean(testAcc[-100:])
```

```python
# plot the experiment results
fig,ax = plt.subplots(1,2,figsize=(15,5))

# Plot 1: Average Accuracy vs Dropout Rate
ax[0].plot(dropoutRates,results,'o-')
ax[0].set_xlabel('Dropout proportion')
ax[0].set_ylabel('Average accuracy')
ax[0].legend(['Train','Test'])

# Plot 2: The Difference (Generalization Gap)
# Ideally, we want this difference to be close to zero.
ax[1].plot(dropoutRates,-np.diff(results,axis=1),'o-')
ax[1].plot([0,.9],[0,0],'k--')
ax[1].set_xlabel('Dropout proportion')
ax[1].set_ylabel('Train-test difference (acc%)')

plt.show()
```
