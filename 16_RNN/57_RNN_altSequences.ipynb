# 57: Predicting Alternating Sequences with RNNs

## ðŸŽ¯ Objective
In this notebook, we train a **Recurrent Neural Network (RNN)** to predict the next number in a sequence where the signs alternate (+, -, +, -). 

We will generate synthetic data, build an RNN model using PyTorch, train it, and critically evaluate if it has truly learned the underlying alternating pattern or just memorized the training data.

## ðŸ“š Key Concepts
* **Sequence Prediction:** The task of using past data points to predict the next value in a series.
* **Recurrent Neural Network (RNN):** A network architecture that maintains a "hidden state" (memory) to capture patterns over time.
* **Pattern Learning vs. Memorization:** A core challenge in deep learning. We will test this by flipping the signs in the test data to see if the model generalizes.

## 1. Import Libraries

We import PyTorch for building the neural network, NumPy for numerical operations, and Matplotlib for visualizing our sequences and training progress.

```python
### import libraries
import torch
import torch.nn as nn
import numpy as np

# for printing out status reports
import sys

# for data visualization
import matplotlib.pyplot as plt
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('svg')
```

## 2. Generating Sequential Data

We create a synthetic dataset of **50 numbers**. 
* The magnitudes are random (uniform distribution).
* The signs alternate: the 1st is positive, 2nd is negative, 3rd is positive, etc.
This creates a predictable pattern: `+ - + - + - ...`

```python
# import the data
N = 50

data = torch.zeros(N)

for i in range(N):
  # create alternating signs: (-1)**0 = 1, (-1)**1 = -1, etc.
  data[i] = torch.rand(1) * (-1)**i

plt.figure(figsize=(15,4))
plt.plot([-1,N+1],[0,0],'--',color=[.8,.8,.8])
plt.plot(data,'ks-',markerfacecolor='w')
plt.xlim([-1,N+1])
plt.show()
```

## 3. Building the RNN Model

We define a custom class `rnnnet` that inherits from `nn.Module`. 

### Architecture:
1.  **`nn.RNN`:** This layer processes the sequence. It takes input features and updates its hidden state at each time step.
2.  **`nn.Linear`:** This is the "head" of the network. It takes the output from the RNN layer and maps it to a single predicted value.

```python
class rnnnet(nn.Module):
  def __init__(self,input_size,num_hidden,num_layers):
    super().__init__()

    # RNN Layer
    self.rnn = nn.RNN(input_size,num_hidden,num_layers)

    # linear layer for output
    self.out = nn.Linear(num_hidden,1)

  def forward(self, x):

    # run through the RNN layer
    y,hidden = self.rnn(x) # no explicit hidden state initialization (defaults to zeros)

    # and the output (linear) layer
    y = self.out(y)

    return y,hidden
```

### Model Parameters and Instantiation
We set up the model dimensions:
* **`input_size=1`**: We are feeding 1 number at a time.
* **`num_hidden=5`**: The RNN has a memory vector of size 5.
* **`seqlength=9`**: The model will look at a sequence of 9 numbers to predict the 10th.

```python
# network parameters
input_size =  1 # "channels" of data
num_hidden =  5 # breadth of model (number of units in hidden layers)
num_layers =  1 # depth of model (number of "stacks" of hidden layers)
seqlength  =  9 # number of datapoints used for learning in each segment
batchsize  =  1 # Note: the training code is actually hard-coded to organize data into batchsize=1

# create an instance of the model and inspect
net = rnnnet(input_size,num_hidden,num_layers)

X = torch.rand(seqlength,batchsize,input_size)
y,h = net(X)
print(X.shape)
print(y.shape) # note: one output per sequence element; generally, we take the final output to force a "many-to-one" design.
print(h.shape)
```

### Testing the Forward Pass
Before training, it's good practice to pass a snippet of real data through the untrained model to ensure shapes match and the loss function works.

```python
# test the model with some data
somedata = data[:seqlength].view(seqlength,1,1)
y = net(somedata)

# grab the final predicted value from the output (first element of tuple output of net)
finalValue = y[0][-1]

lossfun = nn.MSELoss()
lossfun(finalValue,data[seqlength].view(1,1))
```

## 4. Training the Model

We train the model for **30 epochs**. 

### The Training Loop
We slide a window of size `seqlength` across the data. For each position:
1.  **Input:** A sequence of 9 numbers.
2.  **Target:** The 10th number.
3.  **Forward Pass:** The RNN processes the sequence.
4.  **Loss:** We compare the *final* output of the RNN (prediction for the 10th step) with the target.
5.  **Backprop:** Update weights.

```python
# number of training epochs
numepochs = 30

# create a new instance of the model (and optimizer!)
net = rnnnet(input_size,num_hidden,num_layers)
optimizer = torch.optim.SGD(net.parameters(),lr=.001)



# initialize losses
losses = np.zeros(numepochs)
signaccuracy = np.zeros(numepochs)

# loop over epochs
for epochi in range(numepochs):

  # loop over data segments
  seglosses = []
  segacc    = []
  hidden_state = torch.zeros(num_layers,batchsize,num_hidden) # reset the hidden state on each epoch

  for timei in range(N-seqlength):

    # grab a snippet of data
    X = data[timei:timei+seqlength].view(seqlength,1,1)
    y = data[timei+seqlength].view(1,1)

    # forward pass and loss
    yHat,hidden_state = net(X)
    finalValue = yHat[-1]
    loss = lossfun(finalValue,y) # compare final value of output

    # backprop
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    # loss from this segment
    seglosses.append(loss.item())

    # also get sign accuracy
    truesign = np.sign(torch.squeeze(y).numpy())
    predsign = np.sign(torch.squeeze(finalValue).detach().numpy())
    accuracy = 100*(truesign==predsign)
    segacc.append(accuracy)

  # average losses from this epoch
  losses[epochi] = np.mean(seglosses)
  signaccuracy[epochi] = np.mean(segacc)

  msg = f'Finished epoch {epochi+1}/{numepochs}'
  sys.stdout.write('\r' + msg)
```

### Visualization of Training
We plot the loss and the accuracy (percentage of times the model predicted the correct sign).

```python
## let's see how the model did!

fig,ax = plt.subplots(1,2,figsize=(16,5))

ax[0].plot(losses,'s-')
ax[0].set_xlabel('Epochs')
ax[0].set_ylabel('Loss')
ax[0].set_title('Model loss')

ax[1].plot(signaccuracy,'m^-',markerfacecolor='g',markersize=15)
ax[1].set_xlabel('Epochs')
ax[1].set_ylabel('Accuracy')
ax[1].set_title('Sign accuracy (final accuracy: %.2f%%)'%signaccuracy[-1])

plt.show()
```

## 5. Evaluation on Training Data

Now we pass the entire training sequence through the model to visualize its predictions. Since we trained on this data, we expect the performance to be good.

```python
# initialize hidden state
h = np.zeros((N,num_hidden))

# initialize predicted values
yHat = np.zeros(N)
yHat[:] = np.nan


# loop over time segments
for timei in range(N-seqlength):

  # grab a snippet of data
  X = data[timei:timei+seqlength].view(seqlength,1,1)
  
  # forward pass and loss
  yy,hh = net(X)
  yHat[timei+seqlength] = yy[-1]
  h[timei+seqlength,:] = hh.detach()


## compute sign-accuracy
truesign = np.sign(data.numpy())
predsign = np.sign(yHat)
signaccuracy = 100*np.mean(truesign[seqlength:]==predsign[seqlength:])


## plot!
fig,ax = plt.subplots(1,3,figsize=(16,4))
ax[0].plot(data,'bs-',label='Actual data')
ax[0].plot(yHat,'ro-',label='Predicted')
ax[0].set_ylim([-1.1,1.1])
ax[0].set_title('Sign accuracy = %.2f%%' %signaccuracy)
ax[0].legend()

ax[1].plot(data-yHat,'k^')
ax[1].set_ylim([-1.1,1.1])

ax[2].plot(data[seqlength:],yHat[seqlength:],'mo')
ax[2].set_xlabel('Real data')
ax[2].set_ylabel('Predicted data')
r = np.corrcoef(data[seqlength:],yHat[seqlength:])
ax[2].set_title(f"r={r[0,1]:.2f} (but Simpson's paradox!)")

plt.suptitle('Performance on training data',fontweight='bold',fontsize=20,y=1.1)
plt.tight_layout()
plt.show()
```

### Inspecting Hidden States
We can peek inside the RNN's "brain" by plotting the activations of the hidden units over time. We might see them oscillating, reflecting the alternating nature of the input.

```python
# show the hidden "states" (units activations)
plt.figure(figsize=(16,5))

plt.plot(h,'s-')
plt.xlabel('Sequence index')
plt.ylabel('State value (a.u.)')
plt.title('Each line is a different hidden unit')
plt.show()
```

## 6. Generalization Test (The Real Challenge)

Did the model learn "alternate signs" or did it just memorize the training sequence? 

We create a **new dataset** where the pattern is flipped: instead of starting with `+`, we start with `-`. If the model truly learned the rule, it should handle this easily. If it memorized the specific values, it will fail.

```python
# Create new data WITH FLIPPED SIGNS!
newdata = torch.zeros(N)
for i in range(N):
  newdata[i] = torch.rand(1) * (-1)**(i+1) # Note the (i+1) here to flip the sign
```

```python
## now test the network!
# note: no learning here!

h = np.zeros((N,num_hidden))

yHat = np.zeros(N)
for timei in range(N-seqlength):

  # grab a snippet of data
  X = newdata[timei:timei+seqlength].view(seqlength,1,1)
  
  # forward pass and loss
  yy,hh = net(X)
  yHat[timei+seqlength] = yy[-1]
  h[timei+seqlength,:] = hh.detach()


# compute sign-accuracy
truesign = np.sign(newdata.numpy())
predsign = np.sign(yHat)
signaccuracy = 100*np.mean(truesign[seqlength:]==predsign[seqlength:])


fig,ax = plt.subplots(1,2,figsize=(18,4))
ax[0].plot(newdata,'bs-',label='Actual data')
ax[0].plot(yHat,'ro-',label='Predicted')
ax[0].set_ylim([-1.1,1.1])
ax[0].legend()

ax[1].plot(newdata-yHat,'k^')
ax[1].set_ylim([-1.1,1.1])
ax[1].set_title('Sign accuracy = %.2f%%' %signaccuracy)

plt.suptitle('Performance on unseen test data',fontweight='bold',fontsize=20,y=1.1)
plt.show()
```
